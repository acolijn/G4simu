```python
from G4Plotter import *
from pandas import DataFrame, read_csv

# General syntax to import a library but no functions: 
##import (library) as (give the library a nickname/alias)
import matplotlib.pyplot as plt
import pandas as pd #this is how I usually import pandas
import sys #only needed to determine Python version number

# Enable inline plotting
%matplotlib inline
```

```python
# GEANT4 output filename
#fname = "../share/G4simu/macros/events_G4QGSP.root"
fname = "../macros/events_22Na_rev1.root"
sourceName = "$^{22}$Na"
A = 10e3 # source activity in Bq
```

```python
f = TFile(fname)
d = f.GetDirectory("events")
t = d.FindObjectAny("events")
```

```python
# retrieve the geometry information from the ROOT output
gdir = f.GetDirectory("detector/geometry")

geometry = {}
for key in gdir.GetListOfKeys():
    parname = key.GetName()
    geometry[parname] = gdir.FindObjectAny(parname).GetVal()
#geometry
```

```python
#
# crunch the GEANT4 ntuple data. make a PANDAS dataframe for further analysis.
#
data = []
nSimulated = d.FindObjectAny("nbevents").GetVal()

n=0
for event in t:
    # event header information
    data.append({
        'ev':  event.eventid, 
        'id' : -1,
        'xp' : event.xp_pri,
        'yp' : event.yp_pri,
        'zp' : event.zp_pri,
        'e'  : event.etot
    })
    # event hit information
    for i in range(len(event.xp)):
        data.append({
            'ev':  event.eventid, 
            'id' : event.collid[i],
            'xp' : event.xp[i],
            'yp' : event.yp[i],
            'zp' : event.zp[i],
            'e'  : event.ed[i]
    })
    n = n+1
    if n%100000 == 0:
        print("Processed ",n," events")
#
# make a PANDAS dataframe
#
df = pd.DataFrame(data=data)
```

Energy deposited in the LXe 

```python
deltaT = nSimulated / A # divide the number of generated events by the source activity to get the time
```

```python
# reweigh with the exposure time to get the rate
values  = df.e[(df.id==0) & (df.e>0)].values 
weights = np.array(len(values)*[1/deltaT])
h = plt.hist(values,bins=100,histtype='step',weights=weights)
plt.title('Source: ' + sourceName + '   Activity ' + str(A/1000) +' kBq')
plt.text(1000,max(h[0])*0.9,
         'total rate = '+str(np.sum(h[0]))+ ' Hz \nsource position = '+
         str(geometry['SourceDisk_Position'])+' mm\nNaI position ='+
         str(geometry['NaI_crystal_Position'])+' mm',
         verticalalignment='top')

plt.xlim([0, 2000])

ax = plt.gca()
ax.set_xlabel('Energy deposit in LXe (keV)')
ax.set_ylabel('Rate (Hz)')

fig = plt.gcf()
fig.set_size_inches(7,5)
fig.savefig('LXe_untriggered.pdf')
```

Energy deposited in the LXe for events in which the NaI detector found a 511keV gamma

```python
# list of events that have a trigger in the NaI (df.id==1)
lst = df.ev[(df.id==1) & (abs(df.e-511)<10) ]
# select these events in the LXe (df.id==0)
df_trig = df[(df.id==0) & (df.ev.isin(lst))]

# get the energy of these events
values = df_trig.e.values
weights = np.array(len(values)*[1/deltaT])
h = plt.hist(values,bins=100,histtype='step',weights=weights)
plt.text(1000,max(h[0])*0.9,
         'total rate = '+str(np.sum(h[0]))+ ' Hz \nsource position = '+
         str(geometry['SourceDisk_Position'])+' mm\nNaI position ='+
         str(geometry['NaI_crystal_Position'])+' mm',
         verticalalignment='top')

plt.xlim([0,2000])
plt.title('Triggered by NaI   Source: ' + sourceName + '   Activity ' + str(A/1000) +' kBq')
ax = plt.gca()
ax.set_xlabel('Energy deposit in LXe (keV)')
ax.set_ylabel('Rate (Hz)')

fig = plt.gcf()
fig.set_size_inches(7,5)
fig.savefig('LXe_triggered.pdf')
```

```python
df_all = df[(df.id==0)]

h2 = plt.hist2d(df_all.xp.values,df_all.yp.values,bins=50,weights=df_all.e.values)
fig = plt.gcf()
fig.set_size_inches(7,7)
ax = plt.gca()
ax.set_xlabel('x (mm)')
ax.set_ylabel('y (mm)')

plt.plot(df_trig.xp[abs(df_trig.e-511)<10].values,df_trig.yp[abs(df_trig.e-511)<10].values,'wo')
#plt.plot(df_trig.xp[df_trig.e>0].values,df_trig[df_trig.e>0].yp.values,'gx')
fig.savefig('LXe_hitpattern_xy.pdf')
```
