```python
from G4Plotter import *
%matplotlib inline
from IPython.html.widgets import interactive
from IPython.display import display
```

```python
fname = "../share/G4simu/macros/events_22Na_rev1.root"
```

```python
def frame(material="Water", plottype="gamma", savefig=False):
    if plottype == "gamma":
        plotGamma(fname,material,savefig)
    elif plottype == "neutron":
        plotNeutron(fname,material,savefig)
```

```python
w = interactive(frame,material=getMaterials(fname),plottype=['gamma','neutron'])
display(w)
```
